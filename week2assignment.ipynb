## Introduction

This report analyzes BTC-to-USD daily closing price data for 2023. 
The analysis involves loading the data using NumPy, calculating descriptive statistics, visualizing the data with a plot, and identifying key price points.The report also includes a box-and-whisker plot of daily price changes,arithmetic mean representation, and outlier analysis.

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the BTC-USD data from CSV
rates = np.genfromtxt("btc-usd.csv", delimiter=",")
```

Now let's define a general function which takes input arguments for analysing any quarter.
In this function we compute various statistical measures

```python
def analyze_quarter(rates, start_day, end_day, color, label):
    quarter_rates = rates[start_day-1:end_day]
    quarter_days = np.arange(start_day, end_day+1)
    
    # Computing statistics using in-built numpy functions
    mean = np.mean(quarter_rates)
    minimum = np.min(quarter_rates)
    q1 = np.percentile(quarter_rates, 25)
    median = np.median(quarter_rates)
    q3 = np.percentile(quarter_rates, 75)
    maximum = np.max(quarter_rates)
    std_dev = np.std(quarter_rates)
    iqr = q3 - q1 # Calculating interquartile range
    
    # Printing the calculated statistics
    print(f"{label} Statistics:")
    print(f"Arithmetic Mean: {mean:.2f}")
    print(f"Minimum: {minimum:.2f}")
    print(f"Q1: {q1:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Q3: {q3:.2f}")
    print(f"Maximum: {maximum:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"Interquartile Range: {iqr:.2f}\n")
    
    # Plot BTC-USD closing prices for the quarter
    plt.plot(quarter_days, quarter_rates, color, label=label)
    plt.legend()
    
    # Find lowest and highest price days
    lowest_day = np.argmin(quarter_rates) + start_day
    highest_day = np.argmax(quarter_rates) + start_day
    
    print(f"Lowest price was on day {lowest_day} (${minimum:.2f})")
    print(f"Highest price was on day {highest_day} (${maximum:.2f})\n-------------------------")
    
```

First let's do an analysis for Q3 2023

```python
analyze_quarter(rates, 182, 273, 'r-', 'Q3 2023') #From July 1 - Sep 30
plt.title("BTC to USD Closing Prices in Q3 2023")
plt.xlabel("Day of Year")
plt.ylabel("Price (USD)")
plt.show()
```

```python
#Analysing each quarter by passing the required arguments to the function
analyze_quarter(rates, 1, 90, 'b-', 'Q1 2023')
analyze_quarter(rates, 91, 181, 'g-', 'Q2 2023')
analyze_quarter(rates, 182, 273, 'r-', 'Q3 2023')
analyze_quarter(rates, 274, 365, 'y-', 'Q4 2023')
plt.title("BTC to USD Closing Prices in 2023")
plt.xlabel("Day of Year")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```

### Box-and-Whisker Plot for Q3 2023 Price Increases/Decreases

This section generates a box-and-whisker plot to visualize the distribution of daily price changes in Q3 2023.

```python
# Boxplot for Q3 2023 daily price increases/decreases
q3_diff = np.diff(rates[181:273])

plt.boxplot(q3_diff, vert=False, patch_artist=True, boxprops=dict(facecolor='yellow'))
plt.plot(np.mean(q3_diff), 1, 'gx', markersize=12, label='Mean')
plt.title("Distribution of BTC-to-USD daily price changes in Q3 2023")
plt.xlabel("Price Change (USD)")
plt.legend()
plt.show()
```

### Box-and-Whisker Plot Discussion

The box plot visually represents the distribution of daily price changes in Q3 2023.
The box shows the interquartile range (IQR), the median, and the whiskers indicate the range of the data.
The green 'x' marks the arithmetic mean.

```python
# Count outliers
q1_diff = np.percentile(q3_diff, 25)
q3_difff = np.percentile(q3_diff, 75) #Quartile range for the difference in the daily prices
iqr_diff = q3_difff - q1_diff
lower_bound = q1_diff - 1.5 * iqr_diff
upper_bound = q3_difff + 1.5 * iqr_diff
outliers = q3_diff[(q3_diff < lower_bound) | (q3_diff > upper_bound)]
print(f"Number of outliers in Q3 2023: {len(outliers)}")
```

### Outlier Analysis Discussion

The code calculates and identifies the outliers in the Q3 2023 daily price changes using the IQR method.
Outliers represent unusually large or small price fluctuations, which could indicate significant market events or volatility.

```python
# Boxplot analysis for the entire year (done additionally)
btc_diff = np.diff(rates)
plt.figure(figsize=(8, 4))
plt.boxplot(btc_diff, vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.plot(np.mean(btc_diff), 1, 'rx', markersize=12, label='Mean')
plt.title("Distribution of BTC-to-USD daily price changes in 2023")
plt.xlabel("Price Change (USD)")
plt.legend()
plt.show()
```

```python
# Counting outliers for full year
y1_diff = np.percentile(btc_diff, 25)
y3_diff = np.percentile(btc_diff, 75)
y_iqr_diff = y3_diff - y1_diff
y_lower_bound = y1_diff - 1.5 * y_iqr_diff
y_upper_bound = y3_diff + 1.5 * y_iqr_diff
y_outliers = ((btc_diff < y_lower_bound) | (btc_diff > y_upper_bound)).sum()
print(f"Number of outliers in 2023: {y_outliers}")
```
